```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loaded the data set
data_set = pd.read_csv('healthcare-dataset-stroke-data.csv')
```

```python
# Data Exploration
data_set.info()
data_set.describe()
```

```python
# Count the occurrences of each gender category
gender_counts = data_set['gender'].value_counts()
```

```python
title_font = {'family':'serif','color':'red','size':20}
label_font = {'family':'serif','color':'green','size':10}

plt.title("Distribution of Gender in Healthcare",fontdict=title_font)
plt.xlabel("Gender",fontdict=label_font)
plt.ylabel("Number of Patients",fontdict=label_font)
plt.bar(gender_counts.index, gender_counts.values,width=0.3,color='#87CEEB',ec='#00008B')
plt.show()
```

```python
x = data_set['age']
```

```python
plt.title("Distribution of Age in Healthcare",fontdict=title_font)
plt.xlabel("Age",fontdict=label_font)
plt.ylabel("Number of Patients",fontdict=label_font)
plt.hist(x,color='#87CEEB',ec='#00008B',width=6)
plt.show()
```
